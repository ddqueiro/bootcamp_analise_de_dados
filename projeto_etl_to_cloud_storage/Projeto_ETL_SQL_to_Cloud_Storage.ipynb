```python
connection = mysql.connector.connect(
    host = host,
    user = usuario,
    password=password,
    database=database,
    ) # linhas para criar a conexão com o banco de dados

cursor = connection.cursor() #criando o cursor para apontar para o banco de dados
query = ''' SELECT * FROM Cliente;
''' #Query ou consulta desejada
cursor.execute(query) #execução da consulta
result = cursor.fetchall() #armazenando os resultados na variavel results
cliente = pd.DataFrame(result,columns=cursor.column_names) #Converte a variavel result em um df pandas nomeado de tabelas
display(cliente)
cursor.close()
connection.close()
```
