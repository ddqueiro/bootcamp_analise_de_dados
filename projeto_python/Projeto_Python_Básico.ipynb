```python
'''
Utilize a função input() realizar um cadastro coletando dados do usuário, como nome,
cidade e estado em que reside, idade, escolaridade, cargo e salário.

# Funcionalidades do Sistema: O sistema deve oferecer as seguintes funcionalidades:

1 - Adicionar um novo funcionário ao sistema, solicitando informações como nome, cargo, salário, e-mail, etc.

2- Atualizar as informações de um funcionário existente com base em um identificador único (como um número de identificação ou nome).Exibir
informações de um funcionário específico.


3 - Listar todos os funcionários cadastrados.


4 - Permitir a remoção de um funcionário do sistema.


Observações Finais:

Certifique-se de implementar um sistema de identificação único para cada funcionário.
Valide as entradas do usuário para garantir que as operações sejam realizadas corretamente e sem erros.
Documente adequadamente seu código com comentários explicativos para ajudar na compreensão. Esse projeto permite aos alunos praticar e
consolidar conceitos fundamentais em Python, como operadores, estruturas de controle, estruturas de dados e funções,
enquanto constroem um aplicativo útil para gerenciar informações de funcionários de uma empresa fictícia.

'''
```

```python
#lista para armazenar os funcionarios
funcionarios = []
 #lista para armazenar os funcionarios
funcionarios = []

funcionario1 = {
  "id": 1,
  "nome": "Ana Silva",
  "cidade": "São Paulo",
  "estado": "SP",
  "idade": 34,
  "escolaridade": "Graduação",
  "cargo": "Analista de Marketing",
  "salario": 4500.00,
  "email": "ana.silva@empresa.com"
}

funcionario2 = {
  "id": 2,
  "nome": "Carlos Pereira",
  "cidade": "Rio de Janeiro",
  "estado": "RJ",
  "idade": 41,
  "escolaridade": "Pós-Graduação",
  "cargo": "Gerente de Projetos",
  "salario": 7500.00,
  "email": "carlos.pereira@empresa.com"
}

funcionarios.append(funcionario1)
funcionarios.append(funcionario2)

#gerar um id para cada funcionario
def gerar_id():
  return len(funcionarios) + 1

#funcao para adicionar um novo funcionario
def adicionar_funcionario():
  # Indent the code inside the function using 4 spaces
  id = len(funcionarios) + 1

# Obter o nome do funcionário
#verificando o nome
  while True:
    nome = input('Qual o nome do funcionário? ')
    if nome.strip() and nome.isalpha():  # Verifica se a entrada não é vazia e se não é numero
        break
    print('O nome não pode ser vazio e não pode ser número. Tente novamente, por gentileza.')

#verificando se a cidade não está vazia
  while True:
    cidade = input('Qual a cidade do funcionário? ')
    if cidade.strip() and cidade.replace(' ', '').isalpha():  # Verifica se a entrada não é vazia e contém apenas letras
        break
    print('A cidade não pode estar em branco e deve conter apenas letras. Tente novamente, por gentileza')

  # Obter o estado, garantindo que não esteja vazia
  while True:
    estado = input('Qual o estado do funcionário? ')
    if estado.strip() and estado.replace(' ', '').isalpha():
        break
    print('O Estado não pode estar em branco e deve conter apenas letras. Tente novamente, por gentileza')

  # Obter a idade, garantindo que seja um número inteiro positivo
  while True:
      try:
          idade = int(input("Digite a idade do funcionário: "))
          if idade <= 0:
              raise ValueError("A idade deve ser um número positivo.")
          break
      except ValueError:
          print("Por favor, digite uma idade válida (número inteiro positivo).")

  # Verificando escolaridade com opções
  while True:
    print('Escolha a escolaridade do funcionário:')
    print('1 - Fundamental')
    print('2 - Ensino Médio')
    print('3 - Superior')

    opcao = input('Digite o número correspondente à escolaridade: ')

    if opcao == '1':
        escolaridade = 'Fundamental'
        break
    elif opcao == '2':
        escolaridade = 'Ensino Médio'
        break
    elif opcao == '3':
        escolaridade = 'Superior'
        break
    else:
        print('Opção inválida. Por favor, escolha 1, 2 ou 3.')


  # Obter o cargo, garantindo que não esteja vazio
  while True:
    cargo = input('Qual o cargo do funcionário? ')
    if cargo.strip() and cargo.isalpha():  # Verifica se a entrada não é vazia e contém apenas letras
        break
    print('O cargo não pode ser vazio e deve conter apenas letras.')

  # Obter o salário, garantindo que seja um número float positivo
  while True:
      try:
          salario = float(input("Digite o salário do funcionário: "))
          if salario <= 0:
              raise ValueError("O salário deve ser um número positivo.")
          break
      except ValueError:
          print("Por favor, digite um valor de salário válido (número positivo).")

  # Obter o e-mail, garantindo que seja válido
  while True:
      email = input("Digite o e-mail do funcionário: ").strip()
      if "@" in email and "." in email:
          break
      else:
          print("Por favor, digite um e-mail válido.")

  # Criar um dicionário para armazenar os dados do funcionário
  funcionario = {
      "id": id,
      "nome": nome,
      "cidade": cidade,
      "estado": estado,
      "idade": idade,
      "escolaridade": escolaridade,
      "cargo": cargo,
      "salario": salario,
      "email": email
  }

  # Adicionar o funcionário à lista
  funcionarios.append(funcionario)
  print(f"Funcionário {nome} adicionado com sucesso!\n")
```

```python
# Atualização das Informações de Funcionários

def atualizar_funcionario():
  id_funcionario = int(input('Digite o ID do funcionário que você deseja atualizar:'))
  if id_funcionario in funcionarios:
    funcionario = funcionarios[id_funcionario]

 # Solicitando os novos dados do funcionário

    print(f'Atualizando informações de {funcionario["nome"]}')
    funcionarios[id_funcionario]["nome"] = input('Digite o novo nome:')
    funcionario["cargo"] = input('Digite o novo cargo:')
    funcionario["cidade"] = input('Digite a nova cidade:')
    funcionario["salario"] = float(input('Digite o novo salário:'))
    funcionario["email"] = input('Digite o novo email:')
    funcionario["escolaridade"] = input('Digite a nova escolaridade:')
    funcionario["idade"] = int(input('Digite a nova idade:'))
    funcionario["estado"] = input('Digite o novo estado:')
    print('Informações atualizadas com sucesso!')
  else:
    print('Funcionario não encontrado.')

```

```python
def exibir_funcionario():
    if len(funcionarios) == 0:
        print('Nenhum funcionário cadastrado no sistema.')
    else:
      for i in range(len(funcionarios)):
          funcionario = funcionarios[i]  # Acessa o funcionário pelo índice
          print(f'ID: {funcionario["id"]}, '
                f'Nome: {funcionario["nome"]}, '
                f'Cargo: {funcionario["cargo"]}, '
                f'Salário: R$ {funcionario["salario"]:.2f}, '
                f'Cidade: {funcionario["cidade"]}, '
                f'Estado: {funcionario["estado"]}, '
                f'Idade: {funcionario["idade"]}, '
                f'Escolaridade: {funcionario["escolaridade"]}, '
                f'E-mail: {funcionario["email"]}')
```

```python
# Remover funcionário
def remover_funcionario():
    # Exibir todos os funcionários e seus IDs
    print("Funcionários cadastrados:")
    for funcionario in funcionarios:  # Iterar sobre a lista de funcionários
        print(f'ID: {funcionario["id"]}, Nome: {funcionario["nome"]}')

    # Solicitar o ID do funcionário a ser removido
    while True:
        try:
            id_funcionario = int(input('Digite o ID do funcionário que você deseja remover (ou 0 para sair): '))
            if id_funcionario == 0:
                print('Saindo da remoção.')
                break  # Sai do loop se o usuário digitar 0

            funcionario_encontrado = False

          # Usando um loop for para encontrar e remover o funcionário
            for funcionario in funcionarios:
                if funcionario["id"] == id_funcionario:
                    funcionarios.remove(funcionario)  # Remove o funcionário
                    print(f'Funcionário {funcionario["nome"]} removido com sucesso!')
                    funcionario_encontrado = True  # Marca que o funcionário foi encontrado
                    break  # Sai do loop após remover o funcionário

            if not funcionario_encontrado:
                print('Funcionário não encontrado.')

        except ValueError:
            print('Por favor, insira um ID válido (número inteiro).')
```

```python
# Integração e validação do sistema

def menu():
    while True:
      print('Sistema de Gerenciamento de Funcionários')
      print('1 - Adiconar um novo funcionário')
      print('2 - Atualizar as informações de um funcionário existente')
      print('3 - Listar todos os funcionários cadastrados')
      print('4 - Remover um funcionário do sistema')
      print('5 - Sair')
      opcao = input('Escolha uma opção:')

      if opcao == '1':
        adicionar_funcionario()
      elif opcao == '2':
        atualizar_funcionario()
      elif opcao == '3':
        exibir_funcionario()
      elif opcao == '4':
        remover_funcionario()
      elif opcao == '5':
        print('Saindo do sistema...')
        break
      else:
        print('Opção inválida. Tente novamente.')
```
